# Let's start on the basics. Variables and arrays

### First, import some libraries

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib as pit
```

```python
x= np.array([[50],[60],[80],[100]])
y=np.array([[250],[160],[220],[300]])
```

```python
modelo= LinearRegression()
modelo.fit(x,y)
```

```python
y_pred = modelo.predict(x)

plt.scatter(x, y, color='green', label='Datos originales')
plt.plot(x, y_pred, color='black', label='Línea de Regresión')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regresión Lineal en Tonos Negros y Verdes')
plt.legend()
plt.grid(True)
plt.show()
```

```python
array_1 = np.array(2,4,7,2,8,8,9) # Create an array of N values
array_2 = np.random.randint(0,10,7) # Create an array of random int values
            # with the same lenght as array_1

print(array_1,array_2) # print both array_1 and array_2
print(array_1.shape, array_2.shape) # print both array_1 and array_2 shapes
print(array_1 + array_2) # print summation of array_1 and array_2
print(np.cdot(array_1, array_2))
```

```python
array_1[3] = 70 # Replace the value of array_1 in pos 3 by 70
print(array_1 + array_2) # Print summation again
```

```python
array_3 = array_1 * array_2 # Apply element-wise multiplication among array_1 and array_2
print(array_3) # Print results
```
